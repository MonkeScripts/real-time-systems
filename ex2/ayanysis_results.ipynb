```python
import pandas as pd
import json
import os

# 指定测试用例路径
test_cases = [
    "DRTS_Project-Test-Cases-master/8-unschedulable-test-case",
    "DRTS_Project-Test-Cases-master/9-unschedulable-test-case",
    "DRTS_Project-Test-Cases-master/10-unschedulable-test-case"
]

# 显示每个测试用例的文件内容
for test_case in test_cases:
    print(f"\n{'='*60}")
    print(f"Test Case: {test_case}")
    print(f"{'='*60}\n")
    
    # 读取并显示tasks.csv
    print("tasks.csv:")
    tasks_file = os.path.join(test_case, "tasks.csv")
    if os.path.exists(tasks_file):
        tasks_df = pd.read_csv(tasks_file)
        display(tasks_df)
    else:
        print(f"File not found: {tasks_file}")
    
    # 读取并显示architecture.csv
    print("\narchitecture.csv:")
    arch_file = os.path.join(test_case, "architecture.csv")
    if os.path.exists(arch_file):
        arch_df = pd.read_csv(arch_file)
        display(arch_df)
    else:
        print(f"File not found: {arch_file}")
    
    # 读取并显示budgets.csv
    print("\nbudgets.csv:")
    budgets_file = os.path.join(test_case, "budgets.csv")
    if os.path.exists(budgets_file):
        budgets_df = pd.read_csv(budgets_file)
        display(budgets_df)
    else:
        print(f"File not found: {budgets_file}")
    
    # 读取并显示analysis_results.json
    print("\nanalysis_results.json:")
    results_file = os.path.join(test_case, "analysis_results.json")
    if os.path.exists(results_file):
        with open(results_file, 'r') as f:
            results = json.load(f)
            print(json.dumps(results, indent=2))
    else:
        print(f"File not found: {results_file}")
    
    print("\n" + "="*60)
```
